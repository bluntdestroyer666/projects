```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('EURUSD.csv')
df.head()
```

```python
df = df[::-1]
df = df.reset_index(0, len(df))
```

```python
def status(i):
    
    print('Day: ' + str(i) + 
              ' Close: ' + str(df.loc[i, 'Close']) + 
              ' Signal: ' + str(df.loc[i, 'Signal']) + 
              ' Bal: ' + str(df.loc[i, 'Bal']) + 
              ' Pos: ' + str(df.loc[i, 'Pos']))
```

```python
# Let's try to find pivot points...

for i in range(len(df)):
    
    # If i == 0 there's no position. Set bal to 1.
    
    if i == 0:
        
        df.loc[i, 'Signal'] = 0
        df.loc[i, 'Bal'] = 1
        df.loc[i, 'Pos'] = 'Closed'
        
        status(i)
        
    #  Check if there's a signal.
        
    else:
        
        ### Signal ###
        
        if df['Close'][i-1] < df['Close'][i]:
            
            df.loc[i, 'Signal'] = 1
            
            ### If you're already holding, don't do anything.
            
            if df['Pos'][i-1] == 'Open':
                
                df.loc[i, 'Pos'] = 'Open'
                df.loc[i, 'Bal'] = df.loc[i-1, 'Bal'] # Change this to change the floating balance.
                
            ### If you're not holding, start holding.
            
            else:
                
                df.loc[i, 'Pos'] = 'Open'
                df.loc[i, 'Bal'] = df.loc[i-1, 'Bal'] + 1 # Change this... we only want to change if it's higher and we're already holding.
                
            status(i)
                
        ### No signal ###
        
        ### Update anyway, you might be holding. If there's no signal and you're holding, close.
        
        else:
            
            df.loc[i, 'Signal'] = 0
            df.loc[i, 'Bal'] = df.loc[i-1, 'Bal']
            df.loc[i, 'Pos'] = 'Closed'
            
            status(i)
```
