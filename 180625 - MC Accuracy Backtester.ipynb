```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
nflx = pd.read_csv('FAANG/' + 'NFLX.csv')
```

```python
print(len(nflx))
```

```python
"""Sim settings: """
sims = 10
days = 30

"""Days to test:"""
days_to_test = 400

sim_df = pd.DataFrame()
```

```python
"""Prediction prices list maker loop tester"""

pre_prices = pd.DataFrame()

sd1_pre = []
sd2_pre = []

start_price_list = []
actual = []


""" Input how many days to go back here: """

start_price_num = -500

for i in range(days_to_test):
    
    prices = nflx['Close'][:start_price_num+1]
    returns = prices.pct_change()
    start_price = prices.iloc[-1]
    
    """
    print(prices.head())
    
    print('start price: ' + str(start_price))
    
    print('price (nflx): ' + str(nflx['Close'].iloc[start_price_num]))
    
    print('start +30: ' + str(nflx['Close'].iloc[start_price_num + 30]))
    """
    
    for x in range(sims):
        count = 0
        daily_vol = returns.std()

        price_ser = []

        price = start_price * (1 + np.random.normal(0, daily_vol))

        price_ser.append(price)

        for y in range(days):
            if count == days:
                break
            price = price_ser[count] * (1 + np.random.normal(0, daily_vol))
            price_ser.append(price)
            count += 1
            
        sim_df[x] = price_ser

    sim_mean = (sim_df.iloc[-1].mean())
    sim_std = (sim_df.iloc[-1].std())

    sd1 = sim_mean - sim_std
    sd2 = sim_mean + sim_std
    
    sd1_pre.append(sd1)
    sd2_pre.append(sd2)
    
    start_price_list.append(start_price)
    
    actual.append(nflx['Close'].iloc[start_price_num + days_to_test])
    
    start_price_num = start_price_num + 1
```

```python
pre_prices['start_price_list'] = start_price_list
pre_prices['sd1'] = sd1_pre
pre_prices['sd2'] = sd2_pre
pre_prices['actual_price'] = actual
```

```python
pre_prices['acc_pc_sd1'] = pre_prices['actual_price'] / pre_prices['sd1']
pre_prices['acc_pc_sd2'] = pre_prices['actual_price'] / pre_prices['sd2']
print(pre_prices)
```

```python
print('sd1 accuracy: ')
print(len(pre_prices[pre_prices['acc_pc_sd1'] > 1])/len(pre_prices))
print('sd2 accuracy: ')
print(len(pre_prices[pre_prices['acc_pc_sd2'] < 1])/len(pre_prices))
```
